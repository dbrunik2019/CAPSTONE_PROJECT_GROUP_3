```python
#dependencies 

import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', 500)
```

```python
recipes_df = pd.read_csv('recipeData.csv', encoding='latin-1')
recipes_df.head()
```

```python
recipes_df.describe
```

```python
# Drop columns we don't need
recipes_df.drop(columns=["Name", "URL", "Size(L)", "BoilSize", "BoilTime", "BoilGravity", "Efficiency", "MashThickness", "PitchRate", "PrimaryTemp", "PrimingMethod", "PrimingAmount"], inplace=True)
recipes_df.head()
```

```python
new_columns_recipes_df = recipes_df
```

```python
# Drop the null columns where all values are null
new_columns_recipes_df.dropna(axis='columns', how='all')
new_columns_recipes_df.describe
```

```python
# Drop the null rows
new_columns_recipes_df.dropna()
new_columns_recipes_df.describe
```

```python
new_columns_recipes_df.dtypes
```

```python
# Within the SugarScale column, remove any rows that have the word "Plato" in the SugarScale column.
cleaner_recipes_df = new_columns_recipes_df[new_columns_recipes_df["SugarScale"] != "Plato"].reset_index(drop=True).copy()
```

```python
cleaner_recipes_df.describe()
```

```python
cleaner_recipes_df.SugarScale.value_counts()
```
