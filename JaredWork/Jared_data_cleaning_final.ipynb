```python
#dependencies 

import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', 500)
```

```python
recipes_df = pd.read_csv('recipeData.csv', encoding='latin-1')
recipes_df.head()
```

```python
# Drop columns we don't need
recipes_df.drop(columns=["URL", "Size(L)", "BoilSize", "BoilTime", "BoilGravity", "Efficiency", "MashThickness", "PitchRate", "PrimaryTemp", "PrimingMethod", "PrimingAmount"], inplace=True)
recipes_df.head()
```

```python
new_columns_recipes_df = recipes_df
```

```python
# Drop the null columns where all values are null
new_columns_recipes_df.dropna(axis='columns', how='all')
new_columns_recipes_df
```

```python
# Drop the null rows
new_columns_recipes_df.dropna()
new_columns_recipes_df
```

```python
# Within the SugarScale column, remove any rows that have the word "Plato" in the SugarScale column.
cleaner_recipes_df = new_columns_recipes_df[new_columns_recipes_df["SugarScale"] != "Plato"].reset_index(drop=True).copy()
```

```python
cleaner_recipes_df
```

```python
# Resource #2
```

```python
guidelines_df = pd.read_csv('guidelines.csv')
guidelines_df.head()
```

```python
new_beer_df = guidelines_df.merge(cleaner_recipes_df, how='inner',left_on='Styles',right_on='Style')
```

```python
cleaner_recipes_df.columns
```

```python
new_beer_df
```

```python
new_beer_df=new_beer_df[['BeerID', 'Name', 'Style','Style_Family_ID', 'StyleID_x', 'OG', 'FG', 'ABV', 'IBU', 'Color',
       'SugarScale', 'BrewMethod']]
```

```python
new_beer_df.to_csv("final_recipes_dataset.csv")
```
